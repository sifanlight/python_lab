## Part 1 (Class)

```python
class Set:
    def __init__(self):
        self.mylist = []

    def add(self, x):
        flag = 0
        for i in range(len(self.mylist)):
            if self.mylist[i]==x:
                flag = 1
                break
        if flag == 0:
            self.mylist.append(x)
            
    def delete(self, x):
        flag = 0
        for i in range(len(self.mylist)):
            if self.mylist[i]==x:
                self.mylist.remove(x)
                break
    def __or__(self,b):
        z = Set()
        for i in range(len(self.mylist)):
            z.add(self.mylist[i])      
        for i in range(len(b.mylist)):
            z.add(b.mylist[i])         
        return z
    
    def __and__(self,b):
        z = Set()
        for i in range(len(self.mylist)):
            for j in range(len(b.mylist)):
                if(self.mylist[i]==b.mylist[j]):
                    z.add(b.mylist[j])                
        return z
    def __eq__(self,b):
        out = 1
        self.mylist.sort()
        b.mylist.sort()
        if(len(self.mylist)==len(b.mylist)):
            for i in range(len(self.mylist)):
                if(self.mylist[i]!=b.mylist[i]):
                    out = 0
        else:
            out = 0                         
        return out
```

```python
# init test
a = Set()
print('a:', a.mylist)

# add test
a.add(1)
a.add(3)
a.add(5)
a.add(7)
a.add(9)
a.add(5)
print('a:', a.mylist)

# delete test
b = Set()
b.add(1)
b.add(2)
b.add(3)
b.add(4)
b.add(5)
b.add(6)
print('b:', b.mylist)
b.delete(6)

# and/or test
print('b:', b.mylist)
c = a | b
print('c:', c.mylist)
d = a & b
print('d:', d.mylist)

# equlity test
a = Set()
b = Set()
print('a==b ? :', a == b)

a = Set()
a.add(5)
b = Set()
print('a==b ? :', a == b)

a = Set()
a.add(5)
b = Set()
b.add(5)
print('a==b ? :', a == b)
```
